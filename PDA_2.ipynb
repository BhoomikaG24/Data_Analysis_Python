```python
import matplotlib.pyplot as plt
sub=["ada","dbms","aj","bio","gt"]
bhoomi=[99,88,98,96,89]
bhoo=[89,86,98,98,80]
plt.scatter(sub,bhoomi,color='k',label='bhoomi score',marker='D')
plt.scatter(sub,bhoo,color='green',label='bhoo score',marker='h')
plt.xlabel('subject')
plt.ylabel('scores')
plt.legend()
```
